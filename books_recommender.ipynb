### Creating a recomender system that preducts books to the customer.
#### The recommender system takes into account the books that the user has previously read.

```python
#import libraries and load dataset
import pandas as pd
df = pd.read_csv('BX-Books.csv',on_bad_lines='skip',encoding='latin-1',sep=';')
df.head()
```

```python
df.info()
```

```python
#Preprocessing data
#Remove duplicates
df.drop_duplicates(subset='Book-Title')
```

```python
sample_size = 15000
df = df.sample(n=sample_size, replace=False, random_state=490)

df = df.reset_index()
df = df.drop('index',axis=1)
```

```python
def clean_test(author):
    result = str(author).lower()
    return(result.replace(' ', ''))
```

```python
df['Book-Author'] = df['Book-Author'].apply(clean_test)
```

```python
#Convert book title and publisher to lowr case
df['Book-Title'] = df['Book-Title'].str.lower()
df['Publisher'] = df['Publisher'].str.lower()
```

```python
#Combine all strings
df2 = df.drop(['ISBN','Image-URL-S','Image-URL-M','Image-URL-L', 'Year-Of-Publication'], axis=1)

df2['data'] = df2[df2.columns[1:]].apply(
    lambda x:  ' '.join(x.dropna().astype(str)), 
    axis=1)
print(df2['data'].head(5))
```

```python
#Vectorize data
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

```python
#Calculate cosine similary
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(vectorized)
print(similarities)
```

```python
df = pd.DataFrame(similarities, columns=df['Book-Title'], index=df['Book-Title']).reset_index()
df.head()
```

```python
#Display user recommendation
input_book = 'fortune\'s flames'
recommendations = pd.DataFrame(df.nlargest(11, input_book)['Book-Title'])
recommendations = recommendations[recommendations['Book-Title']!=input_book]
print(recommendations)
```
